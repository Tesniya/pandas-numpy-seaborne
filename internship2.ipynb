**Logistic** **Regression**

```
y=cs['Credit_Score']
X=cs.drop(['Credit_Score'],axis=1)
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.25)
```

```
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
```

```
log_reg.fit(X_train,y_train)
log_pred=log_reg.predict(X_test)
```

```
log_pred
```

```
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```
confusion_matrix(y_test, log_pred)
```

```
print('Accuracy:', accuracy_score(y_test,log_pred))
print('precision_score:',precision_score(y_test, log_pred, average='weighted'))
print('recall_score:',recall_score(y_test, log_pred, average='weighted'))
print('f1_score:',f1_score(y_test, log_pred, average='weighted'))
```

The accuracy score of 0.5385 indicates that the model is correct in its predictions slightly more than half of the time. However, accuracy alone might not provide the full picture, especially if your dataset is imbalanced.
Precision and recall provide insights into how well the model performs with respect to positive predictions and actual positive instances, respectively.
The F1-score gives a balanced view of precision and recall, useful when you want to compare models or evaluate the overall effectiveness of your classifier.


**Scaling**

```
from sklearn.preprocessing import StandardScaler
```

```
sc=StandardScaler()
sc_Xtrain=sc.fit_transform(X_train)
sc_Xtest = sc.fit_transform(X_test)
```

```
log_reg.fit(sc_Xtrain,y_train)
log_pred=log_reg.predict(sc_Xtest)
```

```
confusion_matrix(y_test, log_pred)
```

```
print('Accuracy:', accuracy_score(y_test,log_pred))
print('precision_score:',precision_score(y_test, log_pred, average='weighted'))
print('recall_score:',recall_score(y_test, log_pred, average='weighted'))
print('f1_score:',f1_score(y_test, log_pred, average='weighted'))
```

These metrics (accuracy, precision, recall, and F1-score) indicate that the model performs well after scaling the data. Scaling typically helps algorithms that rely on distance measures (like K-Nearest Neighbors or SVMs) or gradient descent (like Logistic Regression) to converge faster and perform better.

Consistency: The fact that accuracy, precision, recall, and F1-score are all close in value suggests that your model is making predictions consistently across different metrics.
