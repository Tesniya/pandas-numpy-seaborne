From all the above scatterplots ,the chosen the features: "Alcohol" and "OD280" for clustering

```
x=data[["Alcohol","OD280"]]
```

```
x.columns
```

```
#K Means Clustering#
#Importing K Means
from sklearn.cluster import KMeans
```

```
#Calculating WCSS for different values of clusters
WCSS=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init="k-means++",random_state=42)
  kmeans.fit(x)
  WCSS.append(kmeans.inertia_)
```

```
#Elbow method for choosig the right number of clusters
plt.plot(range(1,11),WCSS)
plt.xlabel('number of clusters')
plt.ylabel("WCSS")
```

from the Elbow method, we have 3 as optimum number of clusters

```
#Fitting the model with 3 clusters
kmeans=KMeans(n_clusters=3,init="k-means++",random_state=42)
```

```
y_kmeans=kmeans.fit_predict(x)
```

```
#Predicted clusters
y_kmeans
```

```
#Plotting the scatterplot of clusters
cluster_centers = pd.DataFrame(kmeans.cluster_centers_, columns =
['Alcohol', 'OD280'])
# Set the figure size
plt.figure(figsize=(10, 8))

#Plotting the Clusters
ax = sns.scatterplot(data = x,
x = 'Alcohol',
y = 'OD280',
hue = kmeans.labels_,
palette = 'colorblind',
alpha = 0.8,
s = 150,
legend = False)

#Plotting the Centroids
ax = sns.scatterplot(data = cluster_centers,
x = 'Alcohol',
y =  'OD280',
hue = cluster_centers.index,
palette = 'colorblind',
s = 600,
marker = 'D',
ec = 'black',
legend = False)

# Adding Centroid Labels
for i in range(len(cluster_centers)):
  plt.text(x = cluster_centers.Alcohol[i],
  y = cluster_centers.OD280[i],
  s = i,
  horizontalalignment='center',
  verticalalignment='center',
  size = 15,
  weight = 'bold',
  color = 'white')
```

Silhouette Score For Valuation of Model

```
from sklearn.metrics import silhouette_score
```

```
#Finding the Silhouette score
silhouette_average=silhouette_score(x,y_kmeans)
silhouette_average
```

```
#Agglomerative Clustering#
#Optimum number of clusters needed using dendrogram method
import scipy.cluster.hierarchy as sch
dend=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title("Dendrogram")
```

from the above,we can see that the vertical distance is maximum for 3 clusters therefore the optimum number of clusters is 3.

```
#Performing Agglomerative Clustering using sklearn
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=3,metric='euclidean',linkage='ward')
y_hc=hc.fit_predict(x)
```

```
#Predicted clusters
y_hc
```

```
#Plotting the Clusters
ax = sns.scatterplot(data = x,
x = 'Alcohol',
y = 'OD280',
hue = hc.labels_,
palette = 'colorblind',
alpha = 0.8,
s = 150,
legend = False)
```

```
from sklearn.metrics import silhouette_score
```

```
silhouette_average=silhouette_score(x,y_hc)
silhouette_average
```

```
#DBSCAN Clustering#
x.head()
```

```
#K-Distance Graph
from sklearn.neighbors import NearestNeighbors

#Here we consider the number of neighbours as 4
neighbors = NearestNeighbors(n_neighbors=4)
neighbors_fit = neighbors.fit(x)
distances, indices = neighbors_fit.kneighbors(x)
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```

optimum value of epsilon is at the point of maximum curvature in the K-Distance Graph, which is almost 0.2 in this case.

```
#Fitting DBSCAN
from sklearn.cluster import DBSCAN
dbscan_opt=DBSCAN(eps=0.2,min_samples=4)
dbscan_opt.fit(x)
```

```
#Predicted clusters
dbscan_opt.fit_predict(x)
```

```
plt.figure(figsize=(10,10))
plt.scatter(x['Alcohol'],x['OD280'],c=dbscan_opt.labels_,s=150,cmap='plasma')
plt.title('DBSCAN Clustering',fontsize=20)
plt.xlabel('Alcohol',fontsize=14)
plt.ylabel('OD280',fontsize=14)
plt.show()
```

```
from sklearn.metrics import silhouette_score
silhouette_average=silhouette_score(x,dbscan_opt.labels_)
silhouette_average
```

Among the 3 clusters, K Means and Agglomerative clustering have maximum silhouette score with comparison to DBSCAN
